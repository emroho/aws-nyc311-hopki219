```python
import pandas as pd
import boto3
import numpy as np

print("Libraries imported successfully")
```

```python
s3 = boto3.client('s3')
response = s3.list_buckets()
bucket_count = len(response['Buckets'])

print(f"Connected to S3")
print(f"Found {bucket_count} S3 buckets")

for bucket in response['Buckets']:
    print(f"  - {bucket['Name']}")
```
